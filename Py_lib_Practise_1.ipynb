## Practice Program 1: Analyzing Penguin Data

For this practice, we will use the **Palmer Penguins dataset**. This dataset contains measurements for different penguin species found in the Palmer Archipelago, Antarctica. It's an excellent dataset for learning data analysis in a biological context, similar to how the Iris dataset is used for botany.

Your task will be to:
1. Load the Palmer Penguins dataset.
2. Inspect the data to understand its structure and identify any missing values.
3. Create a subset of the data.
4. Visualize the distribution of various measurements (like bill length or flipper length) across different penguin species using box and violin plots.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('penguins')
```

```python
diag_1, axes = plt.subplots(2,3, sharey=False, figsize=(17,12),)

sns.boxplot(data=df, x='species', y='bill_length_mm', ax=axes[0,0], palette='viridis', hue='species', legend=True)
sns.boxplot(data=df, x='species', y='bill_depth_mm', ax=axes[0,1], palette='viridis', hue='species', legend=False)
sns.boxplot(data=df, x='species', y='flipper_length_mm', ax=axes[0,2], palette='viridis', hue='species', legend=False)

axes[0,0].set_xlabel('Penguin Species')
axes[0,1].set_xlabel('Penguin Species')
axes[0,2].set_xlabel('Penguin Species')

axes[0,0].set_ylabel('Bill Length(mm)')
axes[0,1].set_ylabel('Bill Depth(mm)')
axes[0,2].set_ylabel('Flipper Length(mm)')

sns.violinplot(data=df, x='species', y='island', ax=axes[1,0], palette='plasma', hue='species', legend=False)
sns.violinplot(data=df, x='species', y='body_mass_g', ax=axes[1,1], palette='plasma', hue='species', legend=True)
sns.violinplot(data=df, x='species', y='sex', ax=axes[1,2], palette='plasma', hue='species', legend=False)

axes[1,0].set_xlabel('Penguin Species')
axes[1,1].set_xlabel('Penguin Species')
axes[1,2].set_xlabel('Penguin Species')

axes[1,0].set_ylabel('Island')
axes[1,1].set_ylabel('Body Mass(g)')
axes[1,2].set_ylabel('Sex')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

```python
sns.set(rc={'figure.figsize':(14,6)})
sns.boxplot(data=df.select_dtypes(include='number'))

plt.title('Boxplot of Numerical Measurements in a Penguin Dataset')
plt.xlabel('Measurable Characteristic')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
