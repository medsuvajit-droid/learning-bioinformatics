# LIBRARIES FOR DATA SCIENCE

1. **Matplotlib**: The 'Canvas'. It provides the foundation for drawing and fine-tuning every detail of a graph.
2. **Seaborn**: The 'Artist'. It uses the Matplotlib canvas to draw complex statistical charts (like Box plots) with very simple commands.
3. **Pandas**: The 'Lab Notebook'. It organizes data into tables (DataFrames), making it easy to filter, sort, and clean your measurements.
4. **NumPy**: The 'Calculator'. It handles the heavy math and large arrays of numbers efficiently behind the scenes.
```

**The Imports:** We are bringing in our toolkit. We use import ... as ... to give them short nicknames so we don't have to type the full name every time.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
````

**Loading Data:** We use Seaborn to load the 'Iris' dataset, a famous collection of measurements for three species of flowers. This gives us a Pandas DataFrame to work with.

````python
df = sns.load_dataset('iris')
````

**Inspecting Data:**  df.head() shows just the top 5 rows, while df shows a larger summary. This is how we 'peek' at our data table.

**Slicing Data:**  We are using Pandas' .iloc to grab a specific 'subset' (rows 5 to 20). It uses Zero Based indexing. This is useful when you only want to look at specific biological samples.

````python
subset = df.iloc[5:21]
````

````python
# Preparing the Layout/Canvas for our Box plot and Violin plot for Outcome vs Sepal Length using Matplotlib
fig, axes = plt.subplots(1, 2, sharey=False, figsize=(10, 4))
# plt.subplots() is a syntax to inform python to create a canvas layout for our graphs. The term "subplots" here is predefined by Matplotlib thus a syntax and not defined by user
# 1, 2 tells Python: 'I want 1 row and 2 columns for my graphs.'
# fig is the whole poster, and axes are the two specific spots (left and right) where we will draw.
# figsize=(10, 4) simply sets the width and height of the poster.
# sharey=False means the x,y axes won't be shared by the box and violin plots

# Box plot illustration using Seaborn
sns.boxplot(x='species', y='sepal_length', data=df, ax=axes[0])
# data=df makes python look for the datas it needs for plotting the graph in the df variable where we loaded the data of iris.
# When you use Seaborn with a Pandas DataFrame, the names you provide for x and y must exactly match...
#..the column headers in your data table (like 'species' or 'sepal_length').
# ax=axes[0]: This tells Seaborn to draw this specifically in the first (left) spot we reserved.

# Violin Plot Illustration using Seaborn
sns.violinplot(x='species', y='sepal_length', data=df, ax=axes[1])
# ax=axes[1]: This tells Seaborn to draw in the second (right) spot.
# Intuition: A violin plot is like a box plot, but the 'fatness' of the shape shows the density...
#..of the data. If the violin is wide at a certain point, it means many of your samples have...
#..that exact measurement. It's great for seeing if your biological data has multiple 'peaks' (bimodal distribution).

plt.show()
# plt.show(): This is the final command that tells Python, 'I'm done drawing, now show me the final result!'
````

````python
# Box plot for all the numerical variables
sns.set(rc={'figure.figsize':(16,5)})
# The whole line sets the rules
# rc stands for 'run command' settings. By setting the figure.figsize here, you are telling..
#...Seaborn: 'Every graph I make from now on should be 16 units wide and 5 units tall.'
# It’s a shortcut so you don't have to specify the size every time you draw.

# Even though 'figure.figsize' is inside single quotes, it is not a user-defined term. It is a predefined configuration key that Matplotlib and Seaborn understand.
#Here is the logic for why it's in quotes:
#The Dictionary Format: In Python, when you see { 'key' : value }, it's called a Dictionary. Dictionaries always require the 'key' (the label) to be a string, which is why the quotes are there.
#The Specific Keyword: Matplotlib has a massive list of settings called 'rcParams' (Run Command Parameters). 'figure.figsize' is the specific, exact name they chose to represent the width and height of a plot.
#Analogy: Imagine you are filling out an official government form. The form has a box labeled "First Name". You didn't invent that label; the government did so they know what information you are providing...
#...In your code, 'figure.figsize' is the label on the form, and (16, 5) is the information you are writing into that box.
#If you changed it to 'my.poster.size', the library wouldn't know what you mean and would simply ignore it or give an error!


# Multiple Box plot illustration
sns.boxplot(data=df.select_dtypes(include='number'))
# This is where you tell the computer to be selective. df.select_dtypes(include='number') is a Pandas command. It scans your iris table and...
#...says, "Keep the columns with measurements (lengths and widths), but hide the 'species' column because it's text."
# sns.boxplot(data=...) then receives this filtered table. Because you didn't specify an x or y, Seaborn's internal logic says, "I see a table with 4 numerical columns...
#...I will draw one box plot for each of them side-by-side so you can compare them instantly."
````
