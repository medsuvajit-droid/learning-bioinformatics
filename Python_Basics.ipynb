# ADVANCE DATA STRUCTURE
1. Arrays
  

*   List: (Mutable), We use [] brackets
*   Dictionary: We use {} brackets
*   tuple: (Non Mutable), We use () brackets





```python
series1 = [10,20,30,40,50,60]
```

```python
print(series1)
```

```python
type(series1)
```

```python
series2 = (10,20,30,40,50,60,70)
```

```python
print(series2)
```

```python
type(series2)
```

In order to retrieve a particular value from the given array we write the index number (which starts from 0) within 3rd bracks **[Zero Based Indexing]**

```python
print(series1[1])
```

If we write Index number within 2nd Bracks, it wont work

```python
print(series1{0})
```

If we want to print values of an array **AFTER** a particular value in that same array we write the index number from and **AFTER** which the values should be printed **FOLLOWED by a semicolon**

```python
print(series1[2:])
```

If we want to print values of an array **BEFORE** a particular value in that same array we write the index number from and **BEFORE** which the values should be printed **AFTER a semicolon**,

*NOTE: Unlike before here the value from whose index is written wont be printed*

```python
print(series1[:2])
```
